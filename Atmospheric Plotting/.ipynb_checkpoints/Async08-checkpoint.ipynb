## Stability Estimates in Earth's Atmosphere

We can get a crude estimate of $ \nabla S $ using a forward difference as 
$$
    \nabla \frac{S_i}{c_p} \approx \frac{1}{\gamma} \frac{\ln T_{i + 1} - \ln T_{i}}{z_{i + 1} - z_i} - \frac{\gamma - 1}{\gamma} \frac{\ln \rho_{i + 1} - \ln \rho_{i}}{z_{i + 1} - z_i} + \mathcal{O}(z_{i + 1} - z_i).
$$

```python
# Required packages and configurations
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%config InlineBackend.figure_format = 'retina'

# Import data
data = pd.read_csv('US_standard_atm_1976.csv', delim_whitespace = True)

N   = data['temp'].size                    # Number of data points
z   = data['alt']                          # Altitudes
lnT = np.log(data['temp'])                 # Log of temeratures
ln = np.log(data['dens'])                 # Log of densities
   = 7/5                                  # Diatomic gas constant

dS  = np.zeros(N - 1)                      # Entropy gradient

# Compute gradients of S
for i in range(N - 1):
    dz = z[i + 1] - z[i]                   # Change in altitude
    dT = lnT[i + 1] - lnT[i]               # Log change in temp
    d = ln[i + 1] - ln[i]               # Log change in density
    
    dS[i] = (dT - ( - 1) * d) / ( * dz) # Change in entropy

# Plot temperature
fig, ax = plt.subplots(figsize = [5,8])
ln1 = ax.plot(data['temp'][1 : N - 1], z[1 : N - 1], label = 'Temp')
ax.set_xlabel('T [K]')
ax.set_ylabel('z [km]')
ax.grid(axis = 'y')

# Plot gradient of entropy
ax2 = ax.twiny()
ln2 = ax2.plot(dS[1 : N], z[1 : N - 1], color = 'xkcd:orange', label = r'$ \nabla S $')
ax2.set_xlabel(r'$ \nabla \frac{S}{c_p} $')

ax.set_title('Entropy Change in Earth\'s Atmosphere')
lns = ln1 + ln2
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc = 'best')
```

From the graph above, we can see that the Earth's atmosphere, under our current assumptions, has a strictly positive $ \nabla S $. Therefore, the Earth is _**stable**_ througout the atmosphere. However, the change in entropy is near zero from 0km to 10km which implies the troposphere is _**near-adiobatic**_.
