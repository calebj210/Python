## US Standard Atmosphere

```python
data = pd.read_csv('US_standard_atm_1976.csv', delim_whitespace=True);

# Plot temperature
fig, ax = plt.subplots(figsize = [5,8])
ax.plot(data['temp'], data['alt'])
ax.set_xlabel('T [K]')
ax.set_ylabel('z [km]')

# Plot density rho
ax2 = ax.twiny()
ax2.plot(data['dens'] / data['dens'][1], data['alt'], color = 'xkcd:orange')
ax2.set_xlabel(r'$ \rho/\rho_0 $ [kg/m$ ^3 $], $ P/P_0 $ [NText/m$^2$]')
ax2.set_xscale('log')

# Plot pressure P
ax2.plot(data['press'] / data['press'][1], data['alt'], color = 'xkcd:green')

ax.set_title('Us Standard Atmosphere')
```

Color note: Temp - blue, density - orange, pressure - green

From the plot above, we can see a few interesting features! Firstly, both pressure and density decrease steadily and at the same rate as altitude increases. In contrast, temperature has a few flucuations as altitude increases due to different dominating effects.

The troposphere is the region with the adiobatic polytrope (denoted by a linear relationship). The stratosphere is given by the region of increasing temperature. The stratosphere absorbs UV radiation by the ozone layer (lower part of the stratosphere) causing the increase in temperature. Finally, the mesosohere is given by the highest region of decreasing temperature.

I would think the troposphere and mesosphere are unstable due to convection as their temperature decreases with height causing hot air to rise. On the flipside, the stratosphere should be stable under convection since the temperature increases with height meaning the hot air is already at the top.

## Model S

```python
# Import solar data
data = pd.read_csv('Model_S.csv', delim_whitespace=True)
N    = len(data['rho'])    # Number of data points

# Plot temperature
fig, ax = plt.subplots()
ax.plot(data['r/R'], data['T'])
ax.set_xlabel('r/R')
ax.set_ylabel('T [K]')
ax.set_yscale('log')

# Plot density rho
ax2 = ax.twinx()
ax2.plot(data['r/R'], data['rho'] / data['rho'][N - 1], color = 'xkcd:orange')
ax2.set_ylabel(r'$ \rho / \rho_0 $ [g/cm$ ^3 $], $ P / P_0 $ [dyn/cm$^2$]')
ax2.set_yscale('log')

# Plot pressure P
ax2.plot(data['r/R'], data['p'] / data['p'][N - 1], color = 'xkcd:green')

ax.set_title('Model S')
```

Color note: Temp - blue, density - orange, pressure - green

Just as in the first plot with the Earth's atmosphere, we see that density and temperature have very similar profiles. As for the temperature, we have two main regions separated by a quick change in slope. The radiative zone is given by a steady (log) decrease in temperature as radiation dominates energy transfer. Moving to the convection zone, the temperature is low enough for some of the ions such as carbon, oxygen, and a few others to keep their electrons. As a result, more of the radiation from the core is absorbed by these heavier ions leading to a faster and faster decrease in temperature as we move into the convection zone. This quick decrease in temperature is a good indicator of the convection zone.

## VAL Atmospheric data

```python
# Import data
data = pd.read_csv('VAL.csv', delim_whitespace=True)

# Plot temperature
fig, ax = plt.subplots()
ax.plot(data['height(km)'], data['T(K)'])
ax.set_xlabel('z [km]')
ax.set_ylabel('T [K]')
ax.set_xscale('log')
ax.set_yscale('log')

# Plot density rho
ax2 = ax.twinx()
ax2.plot(data['height(km)'], data['n_H'] / data['n_H'][129], color = 'xkcd:orange')
ax2.set_ylabel(r'$ \rho / \rho_0 $ [?/?$ ^3 $], $ P / P_0 $ [g/cm$^2$]')
ax2.set_yscale('log')

# Plot pressure P
ax2.plot(data['height(km)'], data['m(g/cm2)'] / data['m(g/cm2)'][129], color = 'xkcd:green')

ax.set_title('VAL Atmosphere')
```

Color note: Temp - blue, density - orange, pressure - green

Just like the other plots, the pressure and density seem to be following the same profile until we get to the end of the chronosphere. At this point, the pressure and density diverge as the atmosphere expands while maintaining pressure. After this divergence, we can see that the pressure and density follow, roughly, shifted profiles of each other.

The temperature curve is very interesting. The tempurature slowly decreases moving along the chronosphere until it hits a little bump and increases a tiny bit. Then, we get to the transition zone which quickly ramps up temperture to the corona.

I would think the atmosphere above the photosphere would generally be unstable due to the sharp temperature gradients leading to solar winds. However when just considering convection, the temperature increases as we move up in the atmosphere which suggests convection would not be occuring and thus we should have stability under convection.
